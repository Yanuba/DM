# Outliers

An outlier is a data point that is very different from most of the remaining data.
Hawkins formally defined the notion of an outlier as follows
> An outlier is an observation which deviates so much from the other observations as to arouse suspicions that it was generated by a different mechanism.

One of the most common method used to found outliers is the Tukey Fence.
This method is based on the interquartile range (IQR):

if $Q_1$ and $Q_3$ are the lower and upper quartile, an outlier is any observation outside the range:

<center> $[ Q_1 - 1.5(Q_3 - Q_1), Q_3+1.5(Q_3-Q_1) ]$ </center>

See also
* Kernel methods
* Neares Neighbor approach
* CLustering approach
*

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 9
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size

def tukey_fence(sequence):
    Q = np.quantile(sequence, [.25,.75])
    Q1 = Q[0]
    Q3 = Q[1]
    IQR = Q3-Q1
    QL = Q1 - 1.5*IQR
    QH = Q3 + 1.5*IQR
    
    return [x for x in sequence if (x < QL or x > QH)]
```

## Time sequence outliers

Considering how many measurements are outliers, in the sense that the time between two measurements is too high or to low wrt the distribution.

One problem is given by the granularity of the timestams, it is in ms. A lot of more values will be considered outliers, also wen close to 10s, so the values are rounded to the closest second

```python
def time_shift(df):
    ts = df.iloc[:,1]

    ts1 = ts[1:].values - ts[0:-1].values
    ts1 = np.concatenate([[0], ts1])
    ts1 = ts1/1000
    ts1 = ts1.round()
    
    ret = df.copy()
    ret[3] = ts1
    
    return ret
```

```python
for path in t_path+s_path+l_path:
    path = os.path.join(root, path)
    df = pd.read_csv(path, header=None)
    df = time_shift(df)
    
    st = df.iloc[0,0].split('/')[-1]
    room = df.iloc[0,0].split('/')[-2]
    
    ts = df.iloc[:,3]
    
    outliers = tukey_fence(ts)
    
    st = df.iloc[0,0].split('/')[-1]
    room = df.iloc[0,0].split('/')[-2]
    
    print(room, st,len(outliers))
```

## Count outliers in time window

Try to count the number of outliers using a sliding time window of x minutes ()

```python
def outliers_slider(df, mins=120, overlap = 0):
        
        value_out = 0
        time_out = 0
        
        delta = np.timedelta64(mins,'m')
        step = delta - np.timedelta64(overlap, 'm')
        
        first_delta = df.iloc[0, 1]
        first_delta = first_delta.astype('datetime64[ms]').astype('datetime64[h]').astype('datetime64[ms]')
        
        window = []
        
        i = 0
        
        start = first_delta - step
        
        while i < len(df):
            
            #build window
            window = []
                        
            start = start + step
            end = start + delta
            end = end.astype('int64')
            
            while i < len(df) and df.iloc[i,1] < end:
                item = df.iloc[i,:]
                window.append(item)
                i += 1  
            
            if len(window) <= 1:
                # get continuative window
                # print('Empty window', start, end)
                continue
                        
            window_df = pd.DataFrame(window)
            value_out += len(tukey_fence(window_df[2]))
            time_out += len(tukey_fence(window_df[2]))
        return value_out, time_out
        
for path in t_path+s_path+l_path:
    path = os.path.join(root, path)
    df = pd.read_csv(path, header=None)
    df = time_shift(df)
    
    st = df.iloc[0,0].split('/')[-1]
    room = df.iloc[0,0].split('/')[-2]
    
    ts = df.iloc[:,3]
    t_outliers = tukey_fence(ts)
    val_outliers = tukey_fence(df.iloc[:,2])
    st = df.iloc[0,0].split('/')[-1]
    room = df.iloc[0,0].split('/')[-2]
    
    print('a',room, st, len(t_outliers), len(val_outliers))
    
    val_outliers, t_outliers = outliers_slider(df)
    print('b',room, st, t_outliers, val_outliers)
```
